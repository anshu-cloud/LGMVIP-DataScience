```python
# importing necessary libraries cv2 and matplotlib for this project
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading the image

```python
# loading  image in code  using cv2 library, then presenting it with help of pylot
img = cv2.imread("/content/king_kohli.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title("Virat Kohli")
plt.imshow(img)
plt.show()
```

# Converting  to Gray Scale

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.title("Gray Image")
plt.imshow(gray_img,cmap='gray')
plt.show()
```

# Inverting the Image

```python
inverted_img= cv2.bitwise_not(gray_img)
plt.imshow(inverted_img,cmap='gray')
plt.show()
```

# Making Image Blur


```python
blur_img = cv2.GaussianBlur(inverted_img,(21,21),sigmaX=0,sigmaY=0)
```

# Pencil Sketch Image


```python
inverted_blur_img =255-blur_img
pencil_sketch =cv2.divide(gray_img,inverted_blur_img,scale = 256.0)
plt.title("Pencil Sketch")
plt.imshow(pencil_sketch,cmap="gray")
plt.show()
```
